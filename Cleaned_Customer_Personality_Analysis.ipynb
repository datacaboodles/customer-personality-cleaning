# Part 2

```python
# .1. Make a copy of the original dataframe
df_clean = df.copy()
```

```python
# 2. Rename all column names to lowercase with underscores
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df.columns.tolist()
```

```python
# 3. Convert 'dt_customer' to datetime format
df_clean['dt_customer'] = pd.to_datetime(df_clean['dt_customer'], dayfirst=True)
```

```python
# 4. Handle missing values in 'income'
# Let's fill them with the median income (robust to outliers)
median_income = df_clean['income'].median()
df_clean['income'].fillna(median_income, inplace=True)
```

```python
# 5. Check unique values in 'education' and 'marital_status' for inconsistencies
unique_education = df_clean['education'].unique()
unique_marital_status = df_clean['marital_status'].unique()
```

```python
# 6. Check income distribution for potential outliers
high_income = df_clean[df_clean['income'] > 200000][['id', 'income']]
```

```python
# Results 
```

# Part 3

```python
# Data Cleaning: Fix
```

```python
# Fix 'marital_status' values by grouping similar or invalid entries
df_clean['marital_status'] = df_clean['marital_status'].replace({
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Single'
})
```

```python
# 2. Group '2n Cycle' in 'education' with 'Master'
df_clean['education'] = df_clean['education'].replace({
    '2n Cycle': 'Master'
})
```

```python
# 3. Remove outlier: Drop rows where income is extremely high (> 200000)
df_clean = df_clean[df_clean['income'] <= 200000]
```

```python
# Final check on unique values and max income
final_marital_status = df_clean['marital_status'].unique()
final_education = df_clean['education'].unique()
max_income = df_clean['income'].max()
```
